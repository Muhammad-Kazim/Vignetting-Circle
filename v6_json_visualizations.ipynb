```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
```

```python
files = os.listdir('v6_based_results')

losses = []
for file in files:
    if file[0] == '1':
        with open(f'v6_based_results/{file}', 'r') as f:
            plt.plot(json.load(f)['TRAINING LOSS'], label=file)
plt.legend()
plt.title('Traning Loss L1')
```
